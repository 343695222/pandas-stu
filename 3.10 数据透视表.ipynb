```python
import numpy as np
import pandas as pd
import seaborn as sns
```

数据透视表(pivot_table)可以认为是groupby的一种便捷方式，可以快速的了解数据

这里的数据是泰坦尼克号的数据

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.pivot_table('survived', index='sex', columns='class') # 按照性别和船舱类型看存活率
```

```python
fare = pd.qcut(titanic['fare'], 3)
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', index=['sex',age], columns=[fare,'class'])
```

```python
titanic.groupby('class')['fare'].mean() # 二、三舱票价相差不是特别大，只有一等舱高很多
```

### 函数签名
DataFrame.pivot_table(data, values=None, index=None, columns=None,
                           aggfunc='mean', fill_value=None, margins=False,
                           dropna=True, margins_name='All')

aggfunc:累计方法（'sum'、'mean'、'count'、'min'、'max' 等），还可以分别制定

titanic.pivot_table(index='sex', columns='class',
                                aggfunc={'survived':sum, 'fare':'mean'})

fill_value 和 dropna 这两个参数 用于处理缺失值
